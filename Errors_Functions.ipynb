### Required Packages

```python
import numpy as np
import matplotlib.pyplot as plt
```

### MSE

```python
x = np.arange(-5, 5, 0.01)
y = np.square(x)
```

```python
plt.plot(x, y, "blue")
plt.grid(True, which="major")
plt.show()
```

```python
def mse(y_pred, y_true):
    return (np.sum((np.subtract(y_pred,y_true))**2)/ len(y_true))
```

```python
print("MSE equals: {}".format(mse(np.array([1,2,3,4,5]), np.array([1.5,2.5,3,5,5]))))
```

### MAE

```python
x = np.arange(-5, 5, 0.01)
y = np.abs(x)
```

```python
plt.plot(x, y, "blue")
plt.grid(True, which="major")
plt.show()
```

```python
def mae(y_pred, y_true):
    return (np.sum(np.abs(np.subtract(y_pred,y_true)))/ len(y_true))
```

```python
print("MAE equals: {}".format(mae(np.array([1,2,3,4,5]), np.array([1.5,2.5,3,5,5]))))
```
